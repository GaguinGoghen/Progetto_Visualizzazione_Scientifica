```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import seaborn as sns

mpl.rcParams["figure.facecolor"]="None"
mpl.rcParams["axes.facecolor"]="None"

df = pd.read_csv('20kSongs.csv')
df
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df2=pd.read_csv('2kSongsVariant.csv')
df2
```

```python
df3=df.merge(df2)
df3.columns
df3 = df3[['Artist', 'Track', 'Album', 'Album_type', 'Danceability', 'Energy',
       'Loudness', 'Speechiness', 'Acousticness', 'Instrumentalness',
       'Liveness', 'Valence', 'Tempo', 'Duration_min','Views', 'Likes', 'Comments', 'Licensed', 'Stream','most_playedon','duration_ms', 'explicit',
       'year', 'popularity', 'key', 'genre']]
df3
```

```python
df3.to_csv("Database.csv")
db = pd.read_csv("database.csv")
```

```python
db = db[db["genre"]!="set()"]
db
```

```python
#------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
```

```python
#Grafico relazione tra esplicità e popolarità

Exp = db[db["explicit"]==True]["popularity"]
NoExp = db[db["explicit"]==False]["popularity"]

fig, ax= plt.subplots()
bh = ax.boxplot([Exp,NoExp],patch_artist=True,notch=True)
for median in bh["medians"]:
    median.set_color("Red")
for patch in bh["boxes"]:
    patch.set_facecolor("white")    
ax.set_xticklabels(['Contenuto Esplicito','Contenuto Non Esplicito'])
plt.ylabel("Popolarità")
plt.ylim(0,100)
plt.show()
```

```python
#In quale modalità le canzoni vengono rilasciate?

pie =db["Album_type"].value_counts()
legenda = ["Album","Singolo","Compilation"]
colori = ["#D2B4DE","#FAD7A0","#ABEBC6"]
plt.rc("figure",figsize=(15.0,10.0))
plt.pie(pie,labels=legenda,colors=colori)
```

```python
#Rapporto tra Danceability e Valence

sns.scatterplot(data = db,x="Valence",y="Danceability")
plt.show()
print(db["Valence"].corr(db["Danceability"]))
```

```python
#Quali artisti hanno gli ascolti più alti a livello globale?
gn = db.groupby(['Artist']).sum()[["Stream"]].sort_values(by="Stream",ascending = False)
gn.head(20)
```

```python
#Quale artista ha la canzone più famosa?

gn = db.groupby(['Artist']).max()[["popularity"]].sort_values(by="popularity",ascending = False)
gn.head(20)
```

```python
#Anni più "danzabili"

gn = db.groupby(['year']).mean()[["Danceability"]].sort_values(by="Danceability",ascending = False)
gn.head(20)
```

```python
#Studio della durata delle canzoni dal 2008 al 2019

gg = db.groupby(["year"]).mean()[['Duration_min']].sort_index()
gg.plot()
plt.xlim([2008,2019])
plt.xlabel(None)
plt.ylabel("Durata in Minuti")
plt.ylim(3,4.2)
plt.xticks(np.arange(2008,2020))
plt.yticks(np.arange(3,4.5,0.5))
plt.legend().remove()
plt.show()
```

```python
#Grafico con tutti gli attributi dal 2008 al 2019

gg = db.groupby(["year"]).mean()[['Instrumentalness']].sort_index()
gh = db.groupby(["year"]).mean()[['Speechiness']].sort_index()
gi = db.groupby(["year"]).mean()[['Valence']].sort_index()
gj = db.groupby(["year"]).mean()[['Danceability']].sort_index()
gk = db.groupby(["year"]).mean()[['Energy']].sort_index()
gl = db.groupby(["year"]).mean()[['Acousticness']].sort_index()

plt.plot(gg,label = "Strumentalità",color = "#F52915",alpha =0.64)
plt.plot(gh,label = "Presenza Di Parole",color = "#1449D6",alpha =0.64)
plt.plot(gi,label = "Indice di Felicità",color = "#E78F2B",alpha =0.64)
plt.plot(gj,label = "Danzabilità",color = "#20C946",alpha =0.64)
plt.plot(gk,label = "Energia",color = "#11F3DF",alpha =0.64)
plt.plot(gl,label = "Acustico",color = "#D014E6",alpha =0.64)
plt.legend()
plt.xlim([2008,2019])
plt.ylim([0,1])
plt.xticks(np.arange(2008,2020))
plt.show()
```
